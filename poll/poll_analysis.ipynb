```python
#label encoding 
lb_f = LabelEncoder()
lb_ftime = LabelEncoder()
lb_fknow = LabelEncoder()

lb_t = LabelEncoder()
lb_ttime = LabelEncoder()
lb_tknow = LabelEncoder()

lb_pollution_fear = LabelEncoder()
lb_pollution_try = LabelEncoder()

new_data = df

food = lb_f.fit_transform(new_data["food"])
food_time = lb_ftime.fit_transform(new_data["food_time"])
food_know = lb_fknow.fit_transform(new_data["food_know"])

transport = lb_t.fit_transform(new_data['transport'])
transport_time = lb_ttime.fit_transform(new_data['transport_time'])
transport_know = lb_tknow.fit_transform(new_data['transport_know'])

pollution_fear  = lb_pollution_fear.fit_transform(new_data['pollution_fear'])
pollution_try  = lb_pollution_try.fit_transform(new_data['pollution_try'])


# drop the original columns and replace them with indicator columns
new_data = new_data.drop(['food','food_time','food_know','transport','transport_time','transport_know', 'pollution_fear', 'pollution_try'], axis = 1)
le_data = pd.DataFrame(new_data)

le_food = pd.DataFrame(food,  columns=['food'])
le_ftime = pd.DataFrame(food_time,  columns=['food_time'])
le_food_know = pd.DataFrame(food_know,  columns=['food_know'])

lb_transport = pd.DataFrame(food,  columns=['transport'])
lb_ttime = pd.DataFrame(food_time,  columns=['transport_time'])
le_transport_know = pd.DataFrame(transport_know, columns=['transport_know'])

le_pollution_fear = pd.DataFrame(pollution_fear,  columns=['pollution_fear'])
le_pollution_try = pd.DataFrame(pollution_try,  columns=['pollution_try'])

le_data = pd.concat((new_data, le_food, le_ftime, le_food_know, lb_transport, lb_ttime, le_transport_know, le_pollution_fear, le_pollution_try), axis = 1)
print(le_data.head)
```
